# 📊  Univariate Analysis

In this step, we analyze each feature individually to understand its distribution, spread, and patterns.

This helps us:
- Identify common user behavior
- Detect outliers
- Understand how values are distributed

### 📌 Distribution of Screen Time

```python
plt.hist(df['daily_screen_time_min'])
plt.title("Distribution of Screen Time")
plt.xlabel("Daily Screen time")
plt.ylabel("Frequency")
plt.savefig("../images/Screen_distribution.png")  
plt.show()
```

### 📊 Screen Time Analysis

- The distribution is approximately normal (bell-shaped), indicating most users have moderate screen time.
- The screen time ranges from around 180 to 600 minutes per day.
- Most users fall within the 300–400 minutes range, showing high daily usage.
- A few users have very low (<200) and very high (>500) screen time, indicating outliers.

💡 Insight:
Most users spend a significant amount of time on screens daily, suggesting that high screen usage is a common behavior.

### 📌 Distribution of Sleep Hours

```python
plt.hist(df['sleep_hours'])
plt.title("Distribution of Sleep Hours")
plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")
plt.savefig("../images/Sleep_distribution.png")
plt.show()
```

### 😴 Sleep Hours Analysis

- The distribution is approximately normal, indicating balanced sleep patterns among users.
- Sleep duration ranges from 3 to 10 hours.
- Most users fall in the 5–8 hours range, with a peak around 6–7 hours.
- There are a few users with very low (<4 hours) and high (>9 hours) sleep durations.

💡 Insight:
Most users maintain moderate sleep, but a portion of users with lower sleep hours may be at risk of negative mental health effects.

### 📌 Distribution of Notification Count

```python
plt.hist(df['notification_count'])
plt.title("Distribution of Notification Count")
plt.xlabel("Notification Count")
plt.ylabel("Frequency")
plt.savefig("../images/notification_distribution.png")
plt.show()
```

### 🔔 Notification Count Analysis

- The distribution is approximately normal but slightly right-skewed.
- Notification count ranges from around 20 to 160.
- Most users fall in the 60–100 range, indicating moderate notification exposure.
- A few users receive very high notifications (>130), suggesting possible outliers.

💡 Insight:
Most users experience a moderate level of notifications, but high notification counts for some users may lead to frequent interruptions and reduced focus.

### 📌 Distribution of Anxiety Level

```python
plt.hist(df['anxiety_level'])
plt.title("Distribution of Anxiety Level")
plt.xlabel("Anxiety Level")
plt.ylabel("Frequency")
plt.savefig("../images/anxiety_distribution.png")
plt.show()
          
```

# 📊  Relationship Analysis

In this step, we analyze relationships between behavioral features and mental health indicators.

This helps us:
- Validate assumptions
- Identify patterns between variables
- Understand cause-effect possibilities (not conclusions yet)

### 📌 Screen Time vs Anxiety Level

In this step, we analyze how screen time affects anxiety levels.



```python
plt.scatter(df['daily_screen_time_min'], df['anxiety_level'])
plt.title("Relationship of Screen Time vs Anxiety Level")
plt.xlabel("Screen Time")
plt.ylabel("Anxiety")
plt.savefig("../images/Relationship_Screen_Anxiety.png")
plt.show()
```

### 🔹 Screen Time vs Anxiety Level Analysis

The scatter plot shows a weak to moderate positive relationship between screen time and anxiety levels.

As screen time increases, anxiety tends to increase. However, the data points are widely scattered, indicating that screen time alone does not strongly determine anxiety.

This suggests that other factors, such as sleep and notifications, also influence anxiety levels.

### 📌 Sleep Hours vs Anxiety Level

In this step, we analyze how sleep duration affects anxiety levels.

Sleep is a critical factor in mental health, and we expect it to have a strong relationship with anxiety.

```python
plt.scatter(df['sleep_hours'], df['anxiety_level'])
plt.title("Relationship of Sleep Hours vs Anxiety Level")
plt.xlabel("Sleep Hours")
plt.ylabel("Anxiety Level")
plt.savefig("../images/Relationship_Sleep_Anxiety.png")
plt.show()
```

### 🔹 Sleep Hours vs Anxiety Level Analysis

The scatter plot shows no strong relationship between sleep hours and anxiety levels.

Anxiety levels remain high across all sleep ranges, indicating that sleep alone does not significantly explain anxiety in this dataset.

This suggests that other factors, such as screen time or notifications, may play a more important role in influencing anxiety.

### 📌 Notification Count vs Focus Score

In this step, we analyze how notification interruptions affect user focus.

Frequent notifications can distract users, so we expect a relationship between notification count and focus score.

```python
plt.scatter(df['notification_count'], df['focus_score'])
plt.title("Relationship of Notification vs Focus")
plt.xlabel("Notification Count")
plt.ylabel("Focus Score")
plt.savefig("../images/Relationship_Notify_Focus.png")
plt.show()
```

### 🔹 Notification Count vs Focus Score Analysis

The scatter plot shows a weak negative relationship between notification count and focus score.

As the number of notifications increases, focus tends to decrease slightly. However, the data points are widely scattered, indicating that notifications alone do not strongly determine focus.

This suggests that multiple factors influence user focus, not just notification interruptions.
